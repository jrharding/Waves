### Superposition and interference

Phase difference is the fraction of a cycle difference between the waves. It is able to tell us weather one wave is ahead/behind/at the same point as another wave. This can be described as an angle phase difference (ϕ)
$$ ϕ = 360° \frac {d}{\lambda} $$
Where d = distance waves are apart, λ= wavelength [1] 

When two objects collide, the motion of the objects changes considerably. Superposition is the phenomena that occurs when two waves collide. The changes that happen when objects collide are much more obvious to us, however, wave phenomena can still be observed. When two waves collide, the resulting wave acts as an addition of the two individual waves. The height of the shape at each any particular x-position or time is the sum of the height of the two separate waves at the same x-position or time. 

Constructive interference will occur when the wave combine to form a wave with greater amplitude than the individual waves. Destructive interference occurs when the waves cancel each other out. [3]

## Matter waves
You may be aware that normal matter is made up of atoms which are in turn comprised of electrons, protons and neutrons. These are typically referred to as subatomic particles and the structure of an atom is commonly depicted in school books as a mini solar system with the electron orbiting the combined nucleus of protons/neutrons.

However, when we look at matter on this very small scale we discover that these ‘particles’ also have a wave nature. According to quantum mechanics, these particles are not localised in space like points but exist instead spread over a region of space - like a wave. This spreading out in space translates into a probability of where the ‘particle’ is.


This wave like behaviour is characterised by a wavelength. This is known as the de Broglie wavelength () and can be characterised by:
$$\lambda = \frac{h}{mv}$$
Where $m$ is the mass and $v$ is the velocity of the particle.

Below is a calculator to input values of mass and velocity to find how the de Broglie wavelength changes depending on these values.

```python
h = 6.62e-34
m = float(input("Mass of particle (in kg)"))            # asks for input from user to determine variable "m"
v = float(input("Velocity of particle (in m/s)"))       # float is used to describe a decimal, this is known as a data type. This means only a float can be inputted. Other data types include string (word) and int (integer)

w = h/(m*v)                                             #calculation, w has been used instead of λ

print("The De Broglie wavelength of a particle of mass", m, "kg  and velocity", v, "m/s  is", w, "m")      # shows the result of the calculation
```

Using a simple model for the allowed vibrations in a box with width $L$ we can estimate these energies using the equation.
$$E = \frac{n^2 h^2}{8m_e L^2}$$
for an electron with mass $m_e \approx 10^{-30}$kg and  $h$ is the quantum of action known as Planck's constant.

```python
# Calculate energy levels for values of 'n'

n = float(input("Enter the energy level"));
m = 9.1e-031;              # Mass of the electron, kg
L = 1e-10;                 #length of the orbit, m
h = 6.63e-034;             # Planck's constant, J-s
e = 1.6e-19                 # electron charge [C]



while n%1 != 0:                                                   #when remainder n/1 is not equal to 0, the following code will run
    print("Energy levels must be an integer, please try again")
    n = float(input("Enter the energy level"));

E = n**2*h**2/(8*m*L**2);
    
print ("The energy of the electron moving in its orbit with n=%d is%4.0feV" %(n, E/e))
```
